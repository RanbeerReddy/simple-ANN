```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
#droping the irrelevant columns
df = df.drop(["RowNumber","CustomerId",	"Surname"], axis=1)
df
```

```python
#encoding
lb = LabelEncoder()
df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
#one hot encoding the geography
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
geo_encoder = ohe.fit_transform(df[['Geography']])
geo_encoder
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(), columns=ohe.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
#combining the encoded data with the real df
df = pd.concat([df.drop("Geography", axis=1), geo_encoded_df], axis=1)
df
```

```python
#saving the encoders and scaler in pickle
with open("label_encoder_gender.pkl", "wb") as file:
    pickle.dump(lb, file)

with open("onehot_encoder_geo.pkl", "wb") as file:
    pickle.dump(ohe, file)
```

```python
#dividing the dataset into the independent and dependent features
X = df.drop("Exited", axis=1)
y = df['Exited']

#spliting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
with open("StandardScaler.pkl", 'wb') as file:
    pickle.dump(scaler, file)
```

ANN 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
#Building the ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)), #H1 connected to the input
    Dense(32, activation='relu'), #h2
    Dense(1, activation='sigmoid') #output layer
])
```

```python
model.summary()
```

```python
##Adam optimizer
import tensorflow
optimizer = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
##compile the model
model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
#SETTING  up the tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard 

log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
#Setting up early stopping
early_stopping_callback = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
```

```python
#training the model
history = model.fit(
    X_train, y_train, validation_data=(X_test, y_test), epochs=100,
    callbacks=(tensorflow_callback, early_stopping_callback)
)
```

```python
model.save("model.h5")
```
